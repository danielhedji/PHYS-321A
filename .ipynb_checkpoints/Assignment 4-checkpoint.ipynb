```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#c
#constants
R = 1000 #in km
one_rev = 2000 #in s
omega1 = 0.18 #deg per sec
fcf =1
omega = omega1 * (2 * np.pi / 360) #deg. per sec to rad. per sec
```

```python
#defining the ODE for a non-interial ref. frame
def dy_dt(y,t,omega,fcf):
    y1, y2, y3, y4 = y
    eps = 1e-12
    if abs(y1) < eps:
        y1 = eps
    dydt = [y2, 2.*omega*y1*y4+y1*(fcf*omega**2+y4**2),\
            y4, -(2.*y2/y1)*(omega+y4)]
    return dydt
```

```python
#initial conditions
y0 = [R, -R*omega, 0., 0.]
#time interval w/ timesteps
t_start = 0.
t_end = 350.
nt = 10000

t = np.linspace(t_start, t_end, nt)
#solves the ODE given above; solves for the angle and 
#radial position
sol = integrate.odeint(dy_dt, y0, t, args=(omega,fcf))
radius = sol[1,0]
k = 0
while abs(radius) < R and k < nt-1:
    k += 1
    radius = sol[k,0]
```

```python
#takes the height solution from sol
radius = sol[:,0]
r = R - radius
q = 0
#converts angle solution to degrees
phi = sol[:,2] * 180 / np.pi
#limits to only positive radius values
while q < len(r):
    if (r[q] < 0.):
        phi = phi[:q]
        r = r[:q]
        break
    q += 1
#plots solution
plt.plot(phi,r, label = "Projectile trajectory", 
        color = "magenta")
plt.title("Projectile height as a function of angle")
plt.xlabel("Anglular position in azimuth (degrees)")
plt.ylabel("Height (km)")
plt.legend()
plt.grid()
```

```python
print("The max height of the projectile is approximately:", round(max(r)), \
      "km")
print("The angular position of the projectile when it lands is " \
      "approximately",  round(phi[-1]), "degrees")
```

```python
#d
t = np.linspace(0,318,1000)
prod = omega * t
diff = 1 - prod
#using equations to find the phi and height
#as a function of time; equations were found from
#Analytic Mechanics (Seventh edition) by Fowles and Cassiday, pge. 212
phi = np.arctan(prod/diff) - prod
r = R * (1 - 2 * prod * diff) ** 0.5
h = R - r
#plots the solution
phi = phi * 180 / np.pi
plt.plot(phi,h, color = "orange")
plt.title("Inertial frame of projectile trajectory")
plt.xlabel("Anglular position in azimuth (degrees)")
plt.ylabel("Height (km)")
plt.grid()
```

```python
print("The max height of the projectile is approximately:", round(max(h)), \
      "km")
print("The angular position of the projectile when it lands is " \
      "approximately",  round(phi[-1]), "degrees")
```

Since the two graphs plot the same trajectory, and the values produced for the maximum height and angle are the exact same in both scenarios, this shows that $\phi$ and $H$ agree in the different reference frames.
